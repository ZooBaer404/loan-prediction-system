```python
import pandas as pd

try:
    data = pd.read_csv("income.csv")
except Exception as e:
    print(e)
```

```python
data.info()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['Age'], data['Income($)'])
plt.show()
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=5)
data["cluster"] = model.fit_predict(data[['Age', 'Income($)']])
print(data)
```

```python
d1 = data[data.cluster == 0]
d2 = data[data.cluster == 1]
d3 = data[data.cluster == 2]
d4 = data[data.cluster == 3]
d5 = data[data.cluster == 4]
plt.scatter(d1['Age'], d1['Income($)'], color='red')
plt.scatter(d2['Age'], d2['Income($)'], color='blue')
plt.scatter(d3['Age'], d3['Income($)'], color='black')
plt.scatter(d4['Age'], d4['Income($)'], color='yellow')
plt.scatter(d5['Age'], d5['Income($)'], color='pink')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data['Age'] = scaler.fit_transform(data[['Age']])
data
```
